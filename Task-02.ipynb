```python
df ['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df['Survived'] = df ['Survived'].astype('category')
```

```python
df['Pclass'].astype('category' )
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates (inplace=True)
```

```python
sns.countplot(x= 'Survived', data=df) 
```

```python
sns.countplot(x='Pclass',hue='Survived',data=df,color='green' )
```

```python
sns.countplot(x='Survived',hue='Sex',data=df,color='red' )
```

```python
sns.countplot(x='Survived',hue='Embarked',data=df,color='Orange' )
```

```python
sns.catplot(x= 'PassengerId', y='Survived', data=df,color='black') 
```

```python
sns.boxplot(x='Pclass', y='Age',data=df) 
```

```python
sns.histplot(df['Age'], kde=True) 
```

```python
sns.violinplot(x='Age',y='Sex',data=df)
```

```python
sns.barplot(x = df['Sex'], y = df['Fare'])
```

```python
sns.heatmap(df.corr(), annot = True, cmap = 'viridis')
```
